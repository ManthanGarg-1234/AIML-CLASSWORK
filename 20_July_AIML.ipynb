### Understanding NumPy Array Properties (1D)

This section demonstrates the various properties of a one-dimensional NumPy array, such as its dimensions, memory usage, total number of elements, data type, and memory address.

```python

numbers = np.array([10, 20, 30, 40])
print("numbers: ",numbers)
print(numbers.ndim)
print(numbers.nbytes)
print(numbers.size)
print(numbers.dtype)
print(numbers.data)
```

### Understanding NumPy Array Properties (2D)

This section extends the previous one by showing the same properties for a two-dimensional NumPy array, illustrating how these properties change with higher dimensions.

```python
# numbers: The array itself
# ndim: Number of dimensions
# nbytes: Total bytes consumed by the array elements
# size: Total number of elements
# dtype: Data type of the elements
# data: Python buffer object pointing to the start of the array's data
```

```python

numbers = np.array( [[10, 20, 30, 40] , [50, 60, 70, 80]] )
print("numbers: ",numbers)
print(numbers.ndim)
print(numbers.nbytes)
print(numbers.size)
print(numbers.dtype)
print(numbers.data)
```

### Element-wise Arithmetic Operations with NumPy Arrays

NumPy arrays allow for efficient element-wise arithmetic operations. This section demonstrates addition, subtraction, division, and multiplication between two arrays of the same shape.

```python
# numbers: The 2D array
# ndim: Number of dimensions (should be 2 for a 2D array)
# nbytes: Total bytes consumed by the array elements
# size: Total number of elements (rows * columns)
# dtype: Data type of the elements
# data: Python buffer object pointing to the start of the array's data
```

```python

numbers1 = np.array( [[10, 20, 30, 40] , [50, 60, 70, 80]] )
numbers2 = np.array( [[10, 20, 30, 40] , [50, 60, 70, 80]] )
print("addition of both arrays: ",numbers1+numbers2)
print("subtraction of both arrays: ",numbers1-numbers2)
print("division of both arrays: ",numbers1/numbers2)
print("multiplication of both arrays: ",numbers1*numbers2)

```

### Arithmetic Operations on Python Lists vs. NumPy Arrays

This section highlights the difference in behavior when performing arithmetic operations on standard Python lists compared to NumPy arrays. Python lists concatenate with `+` and do not support element-wise `-`, `/`, or `*` for numerical operations directly. To perform element-wise arithmetic, lists must first be converted into NumPy arrays.

```python
# numbers1 and numbers2 are NumPy arrays.
# When arithmetic operations (+, -, /, *) are performed on NumPy arrays, they are applied element-wise.
# For example, numbers1 + numbers2 adds corresponding elements from both arrays.
```

```python

numbers1 =  ([10, 20, 30, 40])
numbers2 = ([50, 60, 70, 80])
print("addition of both arrays: ",numbers1+numbers2)
print("subtraction of both arrays: ",numbers1-numbers2)
print("division of both arrays: ",numbers1/numbers2)
print("multiplication of both arrays: ",numbers1*numbers2)
```

### Statistical Analysis with NumPy

This section demonstrates how to use NumPy functions to calculate various statistics (mean, size, median, standard deviation, maximum, sum, and dot product) on a list of numerical data.

```python
# Original lists that caused TypeError when attempting element-wise subtraction, division, and multiplication.
# Python lists concatenate with '+', but do not support other element-wise arithmetic operations.
# To perform element-wise operations, convert them to NumPy arrays.
```

```python
Marks = [70, 80, 85, 90,95]
print(np.mean(Marks))
print(np.size(Marks))
print(np.median(Marks))
print(np.std(Marks))
print(np.max(Marks))
print(np.sum(Marks))
print(np.dot(Marks,2)) # 2 se all value multiply
```

```python
# Marks: A Python list of numerical data (e.g., student marks).
# np.mean(): Calculates the average of the elements.
# np.size(): Returns the number of elements in the array/list.
# np.median(): Calculates the median of the elements.
# np.std(): Calculates the standard deviation, a measure of data dispersion.
# np.max(): Finds the maximum value in the elements.
# np.sum(): Calculates the sum of all elements.
# np.dot(Marks, 2): In this context, for a 1D array/list and a scalar, it performs scalar multiplication on each element.
```
